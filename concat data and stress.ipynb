```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for i in range(0,61):
    try:
        data=pd.read_csv(f'./final_data/{i+1}_data.csv')
        data['date']=data['date'].apply(lambda x: x.strip())
        data=data.set_index('date')
        stress=pd.read_csv(f'./stress_new/Stress_u{i:02d}.csv')
        stress['date']=stress['date'].apply(lambda x: x.strip())
        stress=stress.set_index('date')
        out=pd.concat([data, stress],axis=1)

        out=out.dropna(subset=['home_time'])
        out.to_csv(f'./final_data_with_stress/{i}.csv')
    except:pass
out
```

```python
return_data=out.groupby('date').sum()
return_data
```
