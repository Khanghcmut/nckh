```python
def plot_location(data,i):
#     start_lat_long=data[data['tnv']==i].iloc[0,8:10]
#     start_lat_long=start_lat_long.values.tolist()
#     move_lat_long=data[data['tnv']==i].iloc[1:,8:10]
#     move_lat_long=move_lat_long.values.tolist()
#     
#     CharminarTopAttractionLats, CharminarTopAttractionLons = zip(*move_lat_long)  
#     apikey = 'AIzaSyAQKpqNTkg5fNYhP5mb5ykAqg-L3sjrMx4' # (your API key here)
#     myGmap = gmplot.GoogleMapPlotter(start_lat_long[0],start_lat_long[1],12,mapikey=apikey)  
#     myGmap.scatter( CharminarTopAttractionLats, CharminarTopAttractionLons, '#FF0000', size = 50, marker = False )  
#     myGmap.plot(CharminarTopAttractionLats, CharminarTopAttractionLons, 'cornflowerblue', edge_width = 3.0)  
#     myGmap.draw(f"./source_map/{i}volunteermap.html")
    print(f"map {i} done")
    
    voldata=data[data['tnv']==i]

    data_localize = local_data(voldata,0,0)
    

    data_localize=data_localize.reset_index(drop=True)
    
    data_localize=data_localize.sort_values(by='time')
    data_localize['time_difference']=data_localize['time'].diff()
    data_localize.to_csv(f"./mapdata_csv_no_filter/{i}_with_adress.csv")
    
    
#     data_localize['count']=data_localize['count'].apply(lambda x: getlocal(x))
#     data_localize=data_localize.dropna()

    
#     print(data_localize['address'].value_counts())
#     data=data_localize
    
#     data=data.reset_index(drop=True)
#     start_lat_long=data[data['tnv']==i].iloc[0,8:10]

#     start_lat_long=start_lat_long.values.tolist()
#     move_lat_long=data[data['tnv']==i].iloc[1:,8:10]
#     move_lat_long=move_lat_long.values.tolist()

#     move_lat_long
#     CharminarTopAttractionLats, CharminarTopAttractionLons = zip(*move_lat_long)  
#     apikey = 'AIzaSyCWnBCEHYuLKhXqmV-WS7IQ3BvIQx_PQJ4' # (your API key here)

#     myGmap = gmplot.GoogleMapPlotter(start_lat_long[0],start_lat_long[1],12,mapikey=apikey)  

#     myGmap.scatter( CharminarTopAttractionLats, CharminarTopAttractionLons, '#FF0000', size = 50, marker = False )  

#     myGmap.plot(CharminarTopAttractionLats, CharminarTopAttractionLons, 'cornflowerblue', edge_width = 3.0)  

#     myGmap.draw(f"./processed_map/{i}volunteermap_after.html") 
#     print(f"processed map {i} done")
def getlocal(count):
    if count<1:
        return np.nan
    else: return count
def local_data(data,loop,return_data):
    while True:
        qtrange=0.5
        lat=data['latitude'].quantile(qtrange, interpolation="nearest")
        
        err=0.00015

        voldatalocal=data[(data['latitude']<err+lat)&(data['latitude']>-err+lat)]

        long=voldatalocal['longitude'].quantile(qtrange, interpolation="nearest")
        voldatalocal=voldatalocal[(voldatalocal['longitude']<err+long)&(voldatalocal['longitude']>-err+long)]
#         print(lat,long)
#         print(voldatalocal.shape[0])
#         print(data.shape[0])
        
        if (not np.isnan(lat)) and (not np.isnan(long)):
            ##
            geolocator = Nominatim(user_agent="Geocoordinateconverter")

            location = geolocator.reverse(f'{lat},{long}',timeout=None)
            print(location.raw['name'])
            voldatalocal['address']=[location.raw['display_name']]*voldatalocal.shape[0]
            voldatalocal['type']=[location.raw['type']]*voldatalocal.shape[0]
            voldatalocal['address_type']=[location.raw['addresstype']]*voldatalocal.shape[0]
            voldatalocal['hall']=[location.raw['name']]*voldatalocal.shape[0]
            voldatalocal['nation']=[location.raw['address'].get('country_code','')]*voldatalocal.shape[0]
            ##
    
            voldatalocal['count']=[voldatalocal.shape[0]]*voldatalocal.shape[0]


        if loop==0:
            return_data=voldatalocal
            loop=1
        else: 
            return_data=pd.concat([return_data,voldatalocal])

        
#         print(return_data)
        if voldatalocal.shape[0]>0: 
            data=data.drop(voldatalocal.index)
        else:
            print("done loop")
            break 
    return return_data    




    
```
