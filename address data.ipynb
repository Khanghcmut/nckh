```python
def to_address(lat,long,i):
    print(lat,"   ", long)
    geolocator = Nominatim(user_agent="Geocoordinateconverter")
#     geolocator=GoogleV3(api_key='AIzaSyCb0U_-yTWwrwNcrsHws8KzdcuQuo7frE4')
#     geolocator=Photon()
    reverse_func = RateLimiter(geolocator.reverse, min_delay_seconds=0.01)
    location = reverse_func(f'{lat},{long}',timeout=None)
    match i:
        case 1 : return location.raw['display_name']
        case 2 : return location.raw['type']
        case 3: return location.raw['addresstype']
        case 4: return location.raw['name']
        case 5: return location.raw['address'].get('country_code','')
data['address']=data.apply(lambda df:to_address(df['latitude'],df['longitude'],1),axis=1)
print('adress done')
data['type']=data.apply(lambda df:to_address(df['latitude'],df['longitude'],2),axis=1)
data['address_type']=data.apply(lambda df:to_address(df['latitude'],df['longitude'],3),axis=1)
data['hall']=data.apply(lambda df:to_address(df['latitude'],df['longitude'],4),axis=1)
data['nation']=data.apply(lambda df:to_address(df['latitude'],df['longitude'],5),axis=1)
```
