```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
import re
```

```python

df=pd.read_json('./dataset/education/class_info.json')
df=df.T
df.periods=df.periods.to_list()
df
```

```python
from datetime import timedelta
def day_learning(ls):
    x = list(map(lambda x: x["day"], ls))
    x.sort()
    return x
def time_learning(ls,n):
    x = list(map(lambda x: (datetime.strptime(x["start"],'%H:%M')
                                             -datetime(1900,1,1)), ls))
    y= list(map(lambda x: datetime.strptime(x["end"],'%H:%M')-datetime(1900,1,1), ls))
    
    x=np.array(x)
    y=np.array(y)
    X_sec=list(map(lambda x: (datetime.strptime(x["start"],'%H:%M')
                                             -datetime(1900,1,1)).total_seconds(), ls))
    Y_sec=list(map(lambda x: (datetime.strptime(x["end"],'%H:%M')
                                             -datetime(1900,1,1)).total_seconds(), ls))
    
    time=np.subtract(Y_sec,X_sec)
    time=time
    x=x.astype(str).tolist()
    y=y.astype(str).tolist()
    
    match n:
        case 1:return time 
        case 2: return x
        case 3:return y
def time_visualize(s):
    return'{:02}:{:02}:\
{:02}'.format(int(s // 3600),int( s % 3600 // 60),int( s % 60))
df["day_learn"]=df["periods"].apply(lambda ls:day_learning(ls))
df["day_go_to_school"]=df["day_learn"].apply(lambda ls:len(ls))
df["time_learn"]=df["periods"].apply(lambda ls:time_learning(ls,1))
df["start_time"]=df["periods"].apply(lambda ls:time_learning(ls,2))
df["end_time"]=df["periods"].apply(lambda ls:time_learning(ls,3))
df["avg_learn_sub"]=df["time_learn"].apply(lambda ls:np.average(ls))
df["visualize_time"]=df["avg_learn_sub"].apply(lambda s: time_visualize(s))
df=df.reset_index(names='class_name')
df
```

```python
def class_field(name):
    n=re.split('\d',name)
    return n[0]
df['class_field']=df['class_name'].apply(lambda x:class_field(x))
df=df[['class_name', 'class_field', 'location', 'periods', 'day_learn', 'day_go_to_school',
       'time_learn', 'start_time', 'end_time', 'avg_learn_sub',
       'visualize_time']]
```
