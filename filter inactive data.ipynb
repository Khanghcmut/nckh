```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./working_gps.csv')
df
```

```python
data=df[['tnv','distance','velocity','accelerate','hour in date','min in date','latitude','longitude','time']]
data.info()
```

```python
def moving(x):
    if x>20: return True
    else: return False
data['moving']=data['distance'].apply(lambda x: moving(x))
data           
        
            
```

```python

inactive_val=[1]
def inactive(df,inactive_val,i):
    if i!=len(df)-1:
        if df['tnv'].iloc[i]!=df['tnv'].iloc[i-1]:
            if df['moving'].iloc[i]==False: return inactive_val.append(1)
            else:return inactive_val.append(2)
        else:
            if df['moving'].iloc[i]==False:
                if df['moving'].iloc[i+1]==False:
                    return inactive_val.append(np.nan)
                else: return inactive_val.append(1)
            else:return inactive_val.append(2)
    else: 
        if df['moving'].iloc[i]==False: return inactive_val.append(1)
        else:return inactive_val.append(2)
for i in range(1,len(data)):
    inactive(data,inactive_val,i) 
print(len(inactive_val))
data['inactive']=inactive_val
data.to_csv('check.csv')
```

```python
time_difference=[0]
def time_inactive(df,i,time_difference):
    hour=df['hour in date'].iloc[i]-df['hour in date'].iloc[i-1]
    if hour<0:
        hour=24+hour
    minute=df['min in date'].iloc[i]-df['min in date'].iloc[i-1]
    time=hour*60+minute
    return time_difference.append(time)
for i in range(1,len(data)):
    time_inactive(data,i,time_difference)
print(len(time_difference))
data['time_difference']=time_difference
```

```python
def filter_data(df,i):
    if df['inactive'].iloc[i]==0:
        if df['time_difference'].iloc[i]>60:
            df['time_difference'].iloc[i]=np.nan
for i in range(len(data)):
    filter_data(data,i)
data=data.dropna()
data=data.reset_index(drop=True)
```
