```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('./working_gps.csv')
df
```

```python
data=df[['tnv','distance','velocity','accelerate','week_date','date','hour in date','min in date','latitude','longitude','time']]
data.info()
```

```python
def moving(x):
    if x>20: return True
    else: return False
data['moving']=data['distance'].apply(lambda x: moving(x))
data           
        
            
```

```python

active_val=[1]
def active(df,active_val,i):
    if i!=len(df)-1:
        if df['tnv'].iloc[i]!=df['tnv'].iloc[i-1]:
            if df['moving'].iloc[i]==False: return active_val.append(0)
            else:return active_val.append(1)
        else:
            if df['moving'].iloc[i]==False:
                if df['moving'].iloc[i+1]==False:
                    return active_val.append(np.nan)
                else: return active_val.append(0)
            else:return active_val.append(1)
    else: 
        if df['moving'].iloc[i]==False: return active_val.append(0)
        else:return active_val.append(1)
for i in range(1,len(data)):
    active(data,active_val,i) 
print(len(active_val))
data['active']=active_val
data.to_csv('check.csv')
```

```python
data=data.dropna()
data.to_csv('check_inactive.csv')
```

```python
data
```

```python
data=data.reset_index(drop=True)
data
```

```python
time_difference=[0]
def time_inactive(df,i,time_difference):
    hour=df['hour in date'].iloc[i]-df['hour in date'].iloc[i-1]
    if hour<0:
        hour=24+hour
    minute=df['min in date'].iloc[i]-df['min in date'].iloc[i-1]
    time=hour*60+minute
    return time_difference.append(time)
for i in range(1,len(data)):
    time_inactive(data,i,time_difference)
print(len(time_difference))
data['time_difference']=time_difference
```

```python
def filter_data(df,i):
    if df['active'].iloc[i]==0:
        if df['time_difference'].iloc[i]>60:
            df['time_difference'].iloc[i]=np.nan
for i in range(len(data)):
    filter_data(data,i)
data_drop_inactive=data.dropna()
data_drop_inactive=data_drop_inactive.reset_index(drop=True)
```
