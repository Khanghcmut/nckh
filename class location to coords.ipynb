```python
geolocator=Nominatim(user_agent="Geocoordinateconverter")
loc=geolocator.geocode(' carpenter hall,nh')
print(loc.latitude)
```

```python
def correct_hall_name(loc):
    match loc:
        case 'silsby-rocky': return 'silsby hall'
        case 'baker-berry': return '25 North Main Street, Hanover'
        case 'reed': return 'Reed Hall,NH'
        case 'kemeny': return 'kemeny,NH'
        case 'moore': return 'moore Psychology,NH'
        case 'sudikoff': return 'sudikoff hall'
        case 'steele': return 'steele hall,NH'
        case 'lsb':return 'silsby hall'
        case 'carson': return 'carson hall,NH'
        case 'cumming': return '10 Thayer , Hanover, NH 03755, United States'
        case 'maclean': return 'maclean center,nh'
        case 'vac':return 'Black Family Visual Arts Center'
        case 'batrlett': return 'Dartmouth Park, Hanover, N.H'
        case 'hopkins': return 'hopkins center,nh'
        case 'carpenterhall': return 'carpenter hall'
        

df['location']=df['location'].apply(lambda x: correct_hall_name(x))
```

```python
df
```

```python
def lat_long_hall(location,i):
    loc=geolocator.geocode(location)
    if i==1:
        return loc.latitude
    else: return loc.longitude
df['latitude']=df['location'].apply(lambda loc:lat_long_hall(loc,1))
df['longitude']=df['location'].apply(lambda loc:lat_long_hall(loc,2))
```
