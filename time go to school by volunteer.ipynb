```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_class=pd.read_csv('Class.csv',index_col='tnv')
data_class=data_class.drop(data_class.columns[0],axis=1)
data_class
# data_class=data.rename(columns={'1':'tnv', '1.1':'col'})
```

```python
data_class.iloc[1,:]
```

```python
data_class_loc=pd.read_csv('class_define.csv',index_col='class_name')
```

```python
data_class_loc
```

```python
time_school_tnv=[]
for i in data_class.index:
    tnv_class=data_class.loc[i,:]
    total_time=np.array([0,0,0,0,0])
    for j in tnv_class:
        try:
            add_time=np.array(eval(data_class_loc.loc[j,'time_go_school']))
        except:
            add_time=[0,0,0,0,0]
        total_time=np.add(total_time,add_time)

#         total_time=total_time/2
        total_time=list(total_time)
    time_school_tnv.append(total_time)
data_class['time_learn_day']=time_school_tnv
```

```python
data_class.to_csv('time_learn_in_day.csv')
```
