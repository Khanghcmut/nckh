```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('./Stress_csv_new/Stress_u01.csv')
data=data.drop(["Unnamed: 0"],axis=1)
data
```

```python
from datetime import datetime
def changedate(x):
    day=datetime.fromtimestamp(x).strftime(" %d/%m/%Y ")
    return day
data['date']=data['time'].apply(lambda time:changedate(time))
```

```python
data.sort_values('time')
```

```python
df=data.groupby("date")
out_data=df[['level',"time"]].mean()
```

```python
df.to_csv("test.csv")
```
