```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('./Stress_csv_new_1/Stress_u01.csv')
data=data.drop(["Unnamed: 0"],axis=1)
data
```

```python
from datetime import datetime
def changedate(x):
    day=datetime.fromtimestamp(x).strftime(" %d:%m:%Y ")
    return day
data['date']=data['time'].apply(lambda time:changedate(time))
```

```python
data.sort_values('time')
```

```python
df=data.groupby("date")
out_data=df[['level',"time"]].mean()
out_data
```

```python
import os
files=os.listdir('./Stress_csv_new_1')
for file in files:
    if file.endswith('.csv'):
        data=pd.read_csv(f'./Stress_csv_new_1/{file}')
        data=data.drop(["Unnamed: 0"],axis=1)
        data['date']=data['time'].apply(lambda time:changedate(time))
        df=data.groupby("date")
        out_data=df[['level']].mean()
        out_data.to_csv(f'./stress_new/{file}')
```
